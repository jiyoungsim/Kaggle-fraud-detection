```python
# [bid.loc['bidder_id',idx].nunique() for idx in train[bidder_id]]
n_country = bids.groupby('bidder_id')['country'].nunique().reset_index().rename({'country':'n_country'}, axis = 1)
n_country.head()
```

```python
alldata = pd.concat([train,test], sort = False)
```

```python
alldata = alldata.merge(n_country, on = 'bidder_id', how = 'left')
alldata.reset_index(inplace = True, drop = True)
alldata.head()
```

```python
alldata.isnull().sum()
```

```python
alldata.fillna(-1, inplace = True)
```

```python
alldata['payment_account'] = alldata['payment_account'].apply(lambda x: x[:6])
alldata['address'].apply(lambda x: x[:6]).value_counts()[:20]
alldata['address'] = alldata['address'].apply(lambda x: x[:6])
alldata.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
alldata['payment_account'] = le.fit_transform(alldata['payment_account'])
alldata['address'] = le.fit_transform(alldata['address'])
alldata.head()
```

```python
y = train['outcome']
```

```python
alldata.drop(['bidder_id','outcome'], 1, inplace = True)
```

```python
train = alldata[:len(train)]
test = alldata[len(train):]
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 100, random_state = 2019, n_jobs = -1)
rf.fit(train, y)
```

```python
rf_pred = rf.predict_proba(test)
rf_pred
```

```python
# submission = pd.read_csv('/kaggle/input/facebook-recruiting-iv-human-or-bot/sampleSubmission.csv')
submission = pd.read_csv('sampleSubmission.csv')
submission.head()
```

```python
submission['prediction'] = rf_pred[:,1]
submission.to_csv('fraud.csv', index = False)
```
